```python
#%% punto taller 13
texto=input('Introduzca un texto: ')
#Se crea una cadena vacia donde estara el texto sin espacios   
texto1='' 
#Se realiza un ciclo que permitan obtener el texto sin espacios      
for i in range(len(texto)):    
    if (texto[i]!=' '):        
        texto1+=texto[i]
    else:
        texto1+=''
tamano=len(texto1)      #tamaño del vector que contiene el texto sin espacios

a=1     #Estas son  las variable para empezar la secuencia
b=1         
vec=[1]
#Aqui se crea la secuencia que sigue el triangulo [1,3,5,7,.....]
while (b<tamano):
    a=a+2
    b=a+b
    vec.append(a)
filas=len(vec)  #Esta variable nos va a decir el numero de renglones que
                #tendra nuestra piramide
b=0
for i in range(filas):
    b=b+vec[i]
    a=b-vec[i]
    #Esta condicion se cumple cuando se llega a la ultima fila
    if i==filas-1:
     #si es diferente significa que la ultima linea no va ocupar todos los
     #espacios disponibles en la fila  
        if (b-tamano)!=0:   
          # Caso para cuando el numero de letras faltantes en la ultima fila 
          #es menor que el numero de letras en la fila de arriba                         
            if (tamano-a)<vec[i-1]:     
                #C es igual a los espacios que se necesitan para centrar                                   
                c=(vec[i]-(tamano-a))//2
                for k in range(0,c):
                    print(end=" ")
    #Este for imprime los espacios en blancos necesarios. Cuando i esta en la 
    #ultima fila, no va a hacer nada por que el rango superior dara 0                        
    for j in range(0,filas-i-1):
        print(end=" ")
    #Aqui va imprimientos los elementos    
    print(texto1[a:b])
```

```python
#listas (ejercicio 18)
A_nxn = (int)(input('inserte el el tamaño n de la matriz cuadrada: '))#primero
#                              pedimos un numero  entero el cual nos dira el  
#                              tamaño de nuestra matriz cuadrada
A = []#creamos una lista vacia
for fila in range(A_nxn):#en el rango del numero entero
    A.append([0]*A_nxn)#multiplicamos una lista con un cero por el tamaño de la
    #                  matriz para asi crear una sublista con el nuemero de
    #                  elementos igual al numero de columnas de la matriz y esa
    #                  sublista la sumamos (A_nxn)veces a la lista vacia A para
    #                  crear una lista de listas, esto me va a representar una
    #                  matriz cuadrada de tamaño (A_nxn)
for fila in A:
    print(fila)#imprime la forma de la matriz pero la verdadera matriz es A, la
    #         matriz solo tiene cero porque el problema pide una matriz 
    #         triangular superior y estas matrices son matrices con 1 en la 
    #         diagonal de la matriz y en los elementos que quedan arriba de 
    #         siendo los demas elementos igual a 0
for i in range(0,A_nxn):#iteramos en las filas de la matriz
    for j in range(0,A_nxn):#hacemos una iteracion en las columnas de la matriz
        #                    sobre la iteracion en las filas de la maytriz
       if j >= i:#condicionamos para generar la matriz triangular
           A[i][j] = 1#de lo contrario toda la matriz se llenaria de unos
print('la matriz triangula asociada a A es: ', A)#imprime la matriz triangular   
f_max1 = max(A)#encontramos e imprimimos el elemento con mas unos que sera el 
#               elemento mas grande
print('el elemento 1 es la fila con mayor cantida de unos: ',f_max1)#se dice 
#que el elemento 1 es el elemento con mayor cantidad de unos porque una 
#matriz triangular superior la fila 1 va a ser la unica fila que haya unos en
#todas sus columnas
```

```python
#listas ejercicio 20
filas = (int)(input('numero de filas de la matriz: '))#se pide el numero de 
#                                          filas que tendra la matriz     
colum = (int)(input('numero de columnas de la matraz: '))#se pide el numero de 
#                                           de columnas que tendra la matriz

matriz = []#creamos uan lista vacia

for fila in range(filas):#hacemos una iteracion de 0 al numero ed filas
    matriz.append([0]*colum)#multiplicamos una lista con uncero por el 
    #                       numero de columnas para tener una sublista con
    #                       igual elementos al numero de columnas y eso se lo
    #                       sumamos a la lista vacia para obtener la matriz
for fila in matriz:
    print(fila)
    
for i in range(filas):
    a = i#a es el elemento de la fila i
    for j in range(colum):
        b = j#b hace referencia al elemento de la columna j
        matriz[i][j] = (float)(input(f'digite el valor en\
 la pocision {a},{b}: '))#aqui se pidio los numero que llevara cada elemento de 
        #                  de la matriz
print(matriz)# se imprime la matriz
matriz_gira = []#creamos una lista vacias la cual va a ser la matriz girada
for i in range(colum):#para los elementos de las filas ponemos el rango de las
    #                  columnas (intercambiamos)
    matriz_gira.append([])#se crea un sublista que seran las filas de la matriz
    for j in range(filas):#ahora las columnas tendran el rango de las filas
        matriz_gira[i].append(matriz[j][-i-1])#aqui es donde para cada fila 
        #                                      de la nueva matriz se acomodaran
        #                                      los elementos de la columna de 
        #                                       la primera matriz
    print(matriz_gira[i])# se imprime las filas de la matriz para mirar como
    #                      la primera matriz se roto correctamente
```
